```python
import pandas as pd
import partridge as ptg
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
import datetime
import gtfstk as gt

from gtfs_utils import *

alt.renderers.enable('notebook')
alt.data_transformers.enable('json')

sns.set_style("white")
sns.set_context("talk")
sns.set_palette('Set2', 10)
```

```python
gtfs_repo = 'data/gtfs_feeds/'
output_folder = 'data/gtfstk_stats/'
import os
os.listdir(gtfs_repo)
```

```python
date_list = [ '2018-02-01', ]
ts = []
rs = []
for d in date_list:
    feed = gt.read_gtfs(gtfs_repo+d+'.zip', dist_units='m')
    trip_stats = gt.compute_trip_stats(feed)
    ts.append(trip_stats)
    route_stats = feed.compute_route_stats(trip_stats, [d.replace('-', '')], split_directions=True )
    rs.append(route_stats)
```

```python
os.mkdir(output_folder)
```

```python
for d, t, r in zip(date_list, ts, rs):
    t.to_csv(output_folder+d+'_trip_stats.csv')
    r.to_csv(output_folder+d+'_route_stats.csv')
```
